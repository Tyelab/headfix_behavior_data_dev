```python
from pathlib import Path
import pandas as pd
import tables

basepath = Path("Y:/specialk/learned_helplessness/LHE013/twop/")
mergedpath = Path("Y:/specialk/learned_helplessness/LHE013/twop/20210609/20210609_LHE013_plane0_merged.h5")
```

```python
dates = ["20210609"]
```

```python
for date in dates:
    date_path = basepath / date
    blink_glob = date_path.glob("*.csv")
    for i in blink_glob:
        print(i)
        blinks = pd.read_csv(i, header=None)
        blinks.columns = ["frame"]
        blinks["seconds"] = round(blinks.frame / 30, 2)
        print(blinks)
        
        blinks.to_csv(str(date_path / date) + "blinks_seconds.csv")
```
