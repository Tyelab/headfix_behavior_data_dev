```python
from pathlib import Path
import pandas as pd
import tables

basepath = Path("Y:/specialk/learned_helplessness/LHE011/twop/")
mergedpath = Path("Y:/specialk/learned_helplessness/LHE014/twop/20210606/20210606_LHE014_plane0_merged.h5")
```

```python
dates = ["20210608", "20210611", "20210612"]
```

```python
for date in dates:
    date_path = basepath / date
    blink_glob = date_path.glob("*.csv")
    for i in blink_glob:
        print(i)
        blinks = pd.read_csv(i, header=None)
        blinks.columns = ["frame"]
        blinks["seconds"] = round(blinks.frame / 30, 2)
        print(blinks)
        
#         blinks.to_csv(str(date_path) + "blinks_seconds.csv")
```
